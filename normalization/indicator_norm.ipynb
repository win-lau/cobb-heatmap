### Normalizing Indicators Using Min-Max Normalization

Uses the normalize_indicator and normalize_indicator_flip functions. The flip function is to align the direction of the indicators with each other (0-1 scale).

Input: cleaned_data
Output: cleaned_normalized_data

```python
import sys
import os

# get project root for python script import
project_root = os.path.abspath("..")
sys.path.append(project_root)

from scripts.normalization import normalize_indicator, normalize_indicator_flip
```
