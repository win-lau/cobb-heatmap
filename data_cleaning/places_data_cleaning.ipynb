### Data Cleaning for PLACES Data

```python
import pandas as pd
```

```python
# for the lack of reliable transportation and obesity data
transp_obesity_data = pd.read_csv("../raw_data/3_PLACES__Local_Data_for_Better_Health,_Census_Tract_Data,_2025_release_20260330.csv")
display(transp_obesity_data)
```

```python
transp_obesity_subset = transp_obesity_data[["LocationName", "Measure", "Data_Value", "TotalPop18plus"]]
display(transp_obesity_subset)
```

```python
# pivot table to have each indicator be a column
transp_obesity_wide = (
    transp_obesity_subset
    .pivot(index=["LocationName", "TotalPop18plus"], columns="Measure", values="Data_Value")
    .reset_index()
)

display(transp_obesity_wide)

transp_obesity_wide.to_csv("../cleaned_data/transp_obesity_data.csv", index=False)
```
