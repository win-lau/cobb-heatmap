### Calculate Modes of Transportation for Commuting to Work Indicator
Based on ACS data for usage of public transportation + walking + carpooling to get to work

```python
import pandas as pd
```

```python
travel_to_work_data = pd.read_csv("../cleaned_data/travel_to_work_data.csv")
display(travel_to_work_data)
```

```python
# add up the counts of alternative transportation modes and calculate percentage based on total workers
travel_to_work_data["alt_transport_count"] = (
    travel_to_work_data["# Car, truck, or van - carpooled to work, 2024"] +
    travel_to_work_data["# Public transportation to work, 2024"] +
    travel_to_work_data["# Walked to work, 2024"]
)

travel_to_work_data["alt_transport_pct"] = (
    travel_to_work_data["alt_transport_count"] / travel_to_work_data["# Workers 16 years and over, 2024"] * 100
)

display(travel_to_work_data)
```

```python
travel_to_work_data = travel_to_work_data.iloc[:, list(range(7)) + [-1]]
display(travel_to_work_data)

travel_to_work_data.to_csv("../cleaned_data/travel_to_work_data_calculated.csv", index=False)
```
