### Data Extraction for EJI Data

```python
import pandas as pd

df = pd.read_csv('../raw_data/17_EJI_2024_Georgia.csv')
```

```python
# filter for Cobb County
cobb_df = df[df['COUNTY'] == 'Cobb County']
display(cobb_df.head())
```

```python
# select relevant columns
# E_MHLTH = % of individuals reporting not good mental health
cobb_df = cobb_df[['GEOID', 'E_MHLTH']]
display(cobb_df)
```

```python
# save cleaned data to csv
cobb_df.to_csv('../cleaned_data/not_good_mental_health_data.csv', index=False)
```
